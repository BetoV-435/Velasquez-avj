The `iris_sklearn` object, loaded from `sklearn.datasets.load_iris()`, is a scikit-learn 'Bunch' object. This object contains the raw data in the `data` attribute, which represents the features matrix.

```python
print("Features matrix (first 5 rows) from iris_sklearn:\n", iris_sklearn.data[:5])
print("\nFeature names from iris_sklearn:\n", iris_sklearn.feature_names)
```

The other `iris` object is a pandas DataFrame loaded using `sns.load_dataset('iris')`. The features in this DataFrame can be accessed by dropping the 'species' column, which is the target variable.

```python
iris_df_features = iris.drop(columns=['species'])
print("Features from iris DataFrame (first 5 rows):\n", iris_df_features.head())
```
