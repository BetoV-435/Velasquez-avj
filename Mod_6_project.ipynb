```python
import numpy as np

df['TaxRate'] = np.abs(df['TaxRate'])

df['TaxRate'].mean()
```

```python
df['Tax'] = df['UnitPrice'] * df['Quantity'] * df['TaxRate'] / 100
```

```python
df['Tax'].mean()
```

```python
df['Extended'] = (df[('UnitPrice')] * df[('Quantity')]) + df['Tax']
```

```python
df['z_score'] = zscore(np.abs(df['Quantity']))
df[(df['z_score'] <= -3) | (df['z_score'] >= 3)]
```

```python
df['z_score_e'] = zscore(np.abs(df['Extended']))
df[(df['z_score_e'] <= -3) | (df['z_score_e'] >= 3)]
```
