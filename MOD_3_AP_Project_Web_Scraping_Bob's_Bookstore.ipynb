```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python

table = soup.find('table')


headers = [header.text.strip() for header in table.find_all('th')]
rows = table.find_all('tr')
data = []
for row in rows[1:]:
    cols = row.find_all('td')
    data.append([col.text.strip() for col in cols])
df = pd.DataFrame(data, columns=headers)
df
```

```python
df.groupby('Author').agg({'Author': 'count'})
```

```python
df.groupby('Topic').agg({'Topic': 'count'})
```

```python

df['Price'] = df['Price'].replace({'\$': ''}, regex=True)
df['Price'] = df['Price'].astype(float)
df
```

```python
df.groupby('Topic').agg({'Price':'mean'})
```

```python
df['Pages'] = df['Pages'].astype(float)
```

```python
df.groupby('Topic').agg({'Pages':'mean'})
```
