```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/companies.csv')
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='state', data=df)
plt.show()
```

```python
jobs = df['Industry'].value_counts().sort_values().reset_index()

jobs
```

```python


plt.figure(figsize=(10,5))
sns.barplot( y='Industry', x= 'count', data=jobs)
plt.show()
```

```python
g_rate = df.groupby('founded', as_index=False).agg({'growth_percentage': 'mean'})
g_rate
```

```python
plt.figure(figsize=(8,5))

plt.plot(g_rate['founded'], g_rate['growth_percentage'])

plt.title('Growth rate by yeara')
plt.xlabel('founded')
plt.ylabel('growth_percentage')
```
